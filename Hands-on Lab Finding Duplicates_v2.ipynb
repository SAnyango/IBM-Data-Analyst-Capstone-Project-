### Task 2: Analysis of Duplicate Response Patterns
1. Identify duplicate response patterns based on selected columns such as MainBranch, Employment, and RemoteWork.
2. Clarify that these represent multiple respondents with identical answers rather than duplicate records. Analyse which other columns frequently share identical values within these response-pattern groups.
   


```python
## Write your code here

pattern_cols = ['MainBranch', 'Employment', 'RemoteWork']
duplicates = df[df.duplicated(subset=pattern_cols, keep=False)]

print("Rows with Identical Response Patterns")
print(duplicates)

```

```python
pattern_groups_df = df[df.duplicated(subset=pattern_cols, keep=False)]

grouped = pattern_groups_df.groupby(pattern_cols)

#Analyzing other columns
other_columns = [col for col in df.columns if col not in pattern_cols]
consistency_results = []

for col in other_columns:
    # Count how many groups have only 1 unique value for this column
    # (Meaning everyone in that group gave the same answer)
    identical_in_group = grouped[col].nunique().eq(1).sum()
    total_groups = grouped.ngroups
    
    consistency_results.append({
        'Column': col,
        'Groups_with_Identical_Values': identical_in_group,
        'Consistency_Percentage': (identical_in_group / total_groups) * 100
    })

#Convert to DataFrame and sort by the highest consistency
analysis_df = pd.DataFrame(consistency_results).sort_values(by='Consistency_Percentage', ascending=False)

# Display the results
print(f"Total Unique Pattern Groups: {total_groups}")
print("-" * 30)
print(analysis_df.head(15)) # Shows the top 15 columns most likely to be identical
```

### Task 3: Visualize Shared Response Patterns
1. Create visualizations to show the distribution of shared response patterns across different categories.
2. Use bar charts or pie charts to represent the distribution of respondents who share identical values for MainBranch, Employment, and RemoteWork, grouped by Country and Employment.


```python

pattern_cols = ['MainBranch', 'Employment', 'RemoteWork']
df_duplicates = df[df.duplicated(subset=pattern_cols, keep=False)]

#Distribution by Country (Bar Chart)

fig1, ax1 = plt.subplots(figsize=(12, 7)) 
country_dist = df_duplicates['Country'].value_counts().sort_values(ascending=False)

ax1.bar(country_dist.index, country_dist.values, color='teal')
ax1.set_title('Duplicate Response Patterns by Country', fontsize=14, pad=15)
ax1.set_ylabel('Count of Respondents')

plt.xticks(rotation=45, ha='right', fontsize=10)
# Manual adjustment: bottom=0.3 gives 30% of space to country names
plt.subplots_adjust(bottom=0.3, top=0.9)
plt.savefig('distribution_by_country.png')
plt.close()

# 3. Distribution by Employment (Pie Chart)
plt.figure(figsize=(8, 8))
emp_dist = df_duplicates['Employment'].value_counts()
plt.pie(emp_dist, labels=emp_dist.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Duplicate Patterns by Employment Type', pad=20)
plt.axis('equal') 
plt.subplots_adjust(top=0.85)
plt.savefig('distribution_by_employment.png')
plt.close()

# 4. Grouped Analysis: Country and Employment (Stacked Bar)
# Increased width (14) and height (8) to handle many bars/legend
fig2, ax2 = plt.subplots(figsize=(14, 8)) 
grouped_data = pd.crosstab(df_duplicates['Country'], df_duplicates['Employment'])
# Sort to make the chart more readable
grouped_data = grouped_data.sort_values(by=grouped_data.columns[0], ascending=False)

grouped_data.plot(kind='bar', stacked=True, ax=ax2)
ax2.set_title('Pattern Groups: Country vs Employment', fontsize=14, pad=20)
ax2.set_ylabel('Number of Respondents')

plt.xticks(rotation=45, ha='right', fontsize=10)
# Crucial: Adjust bottom for labels AND right for the legend
plt.subplots_adjust(bottom=0.3, top=0.9, right=0.85)
# Place legend outside if it's overlapping bars
ax2.legend(title='Employment', bbox_to_anchor=(1.05, 1), loc='upper left')

plt.savefig('grouped_distribution.png', bbox_inches='tight') # bbox_inches catches stray text
plt.close()
```
