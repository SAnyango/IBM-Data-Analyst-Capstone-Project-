### Task 1: Trends in Compensation Over Age Groups


##### 1. Line Chart of Median `ConvertedCompYearly` by Age Group


- Track how the median yearly compensation (ConvertedCompYearly) changes across different age groups.

- Use a line chart to visualize these trends.


```python
df_clean = df.dropna(subset=['Age', 'ConvertedCompYearly'])

# Calculate Medians by Age Group
median_data = df_clean.groupby('Age')['ConvertedCompYearly'].median()

age_order = [
    'Under 18 years old', 
    '18-24 years old', 
    '25-34 years old', 
    '35-44 years old', 
    '45-54 years old', 
    '55-64 years old', 
    '65 years or older'
]
# Reindex the dataframe to match the logical age order
median_data = median_data.reindex(age_order).dropna()

# Create the Stacked Bar Chart
median_data.plot(
    kind='line', 
    figsize=(12, 7), 
    alpha=0.85
)

# Formatting
plt.title('Median ConvertedCompYearly by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Median Yearly Compensation')
plt.xticks(rotation=45)
plt.legend(['ConvertedCompYearly'], loc='upper left')
plt.grid(axis='y', linestyle='--', alpha=0.5)

plt.tight_layout()
plt.show()
```

##### 2. Line Chart of Median `ConvertedCompYearly` for Ages 25 to 45


For a closer look, plot a line chart focusing on the median compensation for respondents between ages 25 and 45.


```python
## Write your code here

df_clean = df.dropna(subset=['Age', 'ConvertedCompYearly'])


age_map = {
    'Under 18 years old': 15, '18-24 years old': 21, '25-34 years old': 30,
    '35-44 years old': 40, '45-54 years old': 50, '55-64 years old': 60,
    '65 years or older': 70
}
df_clean['Age_Numeric'] = df_clean['Age'].map(age_map)

# Drop rows where the essential columns have missing values
df_filtered = df_clean[(df_clean['Age_Numeric'] >= 25) & (df_clean['Age_Numeric'] <= 45)].dropna(subset=['ConvertedCompYearly'])

# Bins for Compensation (ConvertedCompYearly)
bins = [0, 50000, 100000, 150000, 200000, 500000]
labels = ['<$50k', '$50k-100k', '$100k-150k', '$150k-200k', '>$200k']
df_filtered['CompBin'] = pd.cut(df_filtered['ConvertedCompYearly'], bins=bins, labels=labels)

chart_data = df_filtered.groupby('Age_Numeric')['ConvertedCompYearly'].median()

# Plotting
chart_data.plot(kind='line',
                figsize=(12, 7), 
                )

plt.title('Median ConvertedCompYearly for Ages 25 to 45')
plt.xlabel('Compensation Range')
plt.ylabel('Number of Respondents')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

### Task 2: Trends in Job Satisfaction by Experience Level



##### 1. Line Chart of Job Satisfaction (`JobSatPoints_6`) by Experience Level



- Use a column that approximates experience level to analyze how job satisfaction changes with experience.

- If needed, substitute an available experience-related column for `Experience`.


```python
## Write your code here

df['WorkExp'].unique()
```

```python

avg_satisfaction = df.groupby('WorkExp')['JobSatPoints_6'].mean()

#  Create the Line Chart
plt.figure(figsize=(10, 6))
avg_satisfaction.plot(kind='line', marker='o', linestyle='-', color='b')

# Adding titles and labels
plt.title('Job Satisfaction (JobSatPoints_6) by Experience Level')
plt.xlabel('Experience Level')
plt.ylabel('Average Job Satisfaction (Points)')
plt.grid(True, linestyle='--', alpha=0.7)

# Display the plot
plt.tight_layout()
plt.show()
```

### Task 3: Trends in Job Satisfaction and Compensation by Experience


##### 1.Line Chart of Median ConvertedCompYearly Over Experience Level

- This line chart will track how median compensation (`ConvertedCompYearly`) changes with increasing experience.

- Use a column such as `WorkExp` or another relevant experience-related column.


```python
## Write your code here

df_clean = df.dropna(subset=['WorkExp', 'ConvertedCompYearly'])

# Calculate Median
median_data = df_clean.groupby('WorkExp')['ConvertedCompYearly'].median()

#  Create the Line Chart
plt.figure(figsize=(10, 6))
median_data.plot(kind='line', marker='o', linestyle='-', color='b')

# Adding titles and labels
plt.title('Median ConvertedCompYearly Over Experience Level')
plt.xlabel('Experience Level')
plt.ylabel('Yearly Median Compensation')
plt.grid(True, linestyle='--', alpha=0.7)

# Display the plot
plt.tight_layout()
plt.show()
```

##### 2.Line Chart of Job Satisfaction (`JobSatPoints_6`) Across Experience Levels

- Create a line chart to explore trends in job satisfaction (`JobSatPoints_6`) based on experience level.

- This chart will provide insight into how satisfaction correlates with experience over time


```python
## Write your code here

avg_satisfaction = df.groupby('WorkExp')['JobSatPoints_6'].mean()

#  Create the Line Chart
plt.figure(figsize=(10, 6))
avg_satisfaction.plot(kind='line', marker='o', linestyle='-', color='b')

# Adding titles and labels
plt.title('Job Satisfaction (JobSatPoints_6) by Experience Level')
plt.xlabel('Experience Level')
plt.ylabel('Average Job Satisfaction (Points)')
plt.grid(True, linestyle='--', alpha=0.7)

# Display the plot
plt.tight_layout()
plt.show()
```
