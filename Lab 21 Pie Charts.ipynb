### Task 1: Visualizing Data Composition with Pie Charts


##### 1.1 Create a Pie Chart of the Top 5 Databases Respondents Want to Work With


In the survey data, the `DatabaseWantToWorkWith` column lists the databases that respondents wish to work with. Let’s visualize the top 5 most-desired databases in a pie chart.



```python
##Write your code here

df_clean = df.dropna(subset=['DatabaseWantToWorkWith'])

df_counts = df_clean['DatabaseWantToWorkWith'].str.split(';').explode().value_counts()

top_5_db = df_counts.head(5)

# Create the Pie Chart
plt.figure(figsize=(10, 7))
plt.pie(
    top_5_db, 
    labels=top_5_db.index, 
    autopct='%1.1f%%',   # Display percentages
    startangle=140,      # Rotate for better aesthetics
)

# Formatting
plt.title('Top 5 Databases Respondents Want to Work With')
plt.axis('equal')  # Ensure the pie is a circle
plt.show()
```

The `DevType` column lists the developer types for respondents. We’ll examine the distribution by showing the top 5 developer roles in a pie chart.



```python
##Write your code here
df_clean = df.dropna(subset=['DevType'])

df_counts = df_clean['DevType'].str.split(';').explode().value_counts()

top_5_dev = df_counts.head(5)

# Create the Pie Chart
plt.figure(figsize=(10, 7))
plt.pie(
    top_5_dev, 
    labels=top_5_dev.index, 
    autopct='%1.1f%%',   # Display percentages
    startangle=140,      # Rotate for better aesthetics
)

# Formatting
plt.title('Top 5 Developer Types For Respondents')
plt.axis('equal')  # Ensure the pie is a circle
plt.show()
```
